```python
df['children'].fillna(0, inplace=True)
df['agent'].fillna(0, inplace=True)
df['company'].fillna(0, inplace=True)
```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
df.drop(columns=['company'], inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
df['agent'].fillna('Not Provided', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
numeric_cols = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
                'adults', 'children', 'babies', 'adr', 'previous_cancellations', 'previous_bookings_not_canceled']

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df['revenue'] = df['adr'] * (df['stays_in_week_nights'] + df['stays_in_weekend_nights'])
```

```python
df['arrival_date'] = pd.to_datetime(df[['arrival_date_year', 'arrival_date_month']].astype(str).agg('-'.join, axis=1))
```

```python
revenue_trends = df.groupby(df['arrival_date'].dt.to_period("M"))['revenue'].sum()
```

```python
revenue_trends.plot(kind='line', title="Revenue Trends Over Time")
```

```python
cancellation_rate = df['is_canceled'].mean() * 100
print(f"Cancellation Rate: {cancellation_rate:.2f}%")
```

```python
country_counts = df['country'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10])
plt.xticks(rotation=45)
plt.title("Top 10 Booking Countries")
plt.show()
```

```python
sns.histplot(df['lead_time'], bins=50, kde=True)
plt.title("Booking Lead Time Distribution")
plt.show()
```

```python
weekend_bookings = df['stays_in_weekend_nights'].sum()
weekday_bookings = df['stays_in_week_nights'].sum()
```

```python
print(f"Total Weekend Bookings: {weekend_bookings}")
print(f"Total Weekday Bookings: {weekday_bookings}")
```

```python
labels = ['Weekdays', 'Weekends']
values = [weekday_bookings, weekend_bookings]

plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['blue', 'orange'])
plt.title("Weekend vs Weekday Bookings")
plt.xlabel("Type")
plt.ylabel("Number of Nights Booked")
plt.show()
```

```python
popular_room_type = df['reserved_room_type'].value_counts()
```

```python
print("Most Popular Room Types:\n", popular_room_type)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=popular_room_type.index, y=popular_room_type.values, palette='viridis')
plt.title("Most Popular Room Types")
plt.xlabel("Room Type")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
season_mapping = {
    'December': 'High', 'January': 'High', 'February': 'High',  # Winter Holidays
    'March': 'Low', 'April': 'Low', 'May': 'Medium',            # Spring, before peak
    'June': 'High', 'July': 'High', 'August': 'High',           # Summer Vacation
    'September': 'Medium', 'October': 'Medium', 'November': 'Low'  # Fall, pre-holiday lull
}
```

```python
df['season'] = df['arrival_date_month'].map(season_mapping)
```

```python
season_counts = df['season'].value_counts()
```

```python
print("Booking Counts per Season:\n", season_counts)
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x=season_counts.index, y=season_counts.values, palette='coolwarm')
plt.title("High-Season vs Low-Season Booking Trends")
plt.xlabel("Season")
plt.ylabel("Number of Bookings")
plt.show()
```
